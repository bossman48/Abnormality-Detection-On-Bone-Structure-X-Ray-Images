```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

modelLenet5 = keras.models.Sequential([
    keras.layers.Conv2D(6, kernel_size=5, strides=1,  activation='tanh', input_shape=(224,224,3), padding='same'), #C1
    keras.layers.AveragePooling2D(), #S2
    keras.layers.Conv2D(16, kernel_size=5, strides=1, activation='tanh', padding='valid'), #C3
    keras.layers.AveragePooling2D(), #S4
    keras.layers.Conv2D(120, kernel_size=5, strides=1, activation='tanh', padding='valid'), #C5
    keras.layers.Flatten(name="flattenLayer"), #Flatten
    keras.layers.Dense(120, activation='tanh'), #F6    
    keras.layers.Dense(84, activation='tanh'), #F6
    keras.layers.Dense(2, activation='softmax') #Output layer
])

modelLenet5.summary()
```

```python
#model compile parameters is set
modelLenet5.compile(optimizer="adam",
              loss= tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```
